```python
import multiprocessing
from sympy import Point, Polygon
import math
import matplotlib.pyplot as plt
import time
```

```python
a0=1
a1=1

obstacles=(Polygon((-0.25, 1.1), (-0.25, 2), (0.25, 2), (0.25, 1.1)),
           Polygon((-2, -2), (-2, -1.8), (2, -1.8), (2,-2)))

p0=(0,0)
m1=20
m2=20
points=[]

def sum_up_to(theta1):
    for theta2 in range(int(360/m2)):
        p1=((a0*(math.cos(theta1*m1))),
            (a0*(math.sin(theta1*m1))))
        p2=((a0*(math.cos(theta1*m1))+a1*(math.cos(theta1*m1+theta2*m2))),
            (a0*(math.sin(theta1*m1))+a1*(math.sin(theta1*m1+theta2*m2))))
        links=(Polygon(p0,p1),Polygon(p1,p2))
        print(theta1,theta2)
        for obstacle in obstacles:
            for link in links:
                if len(link.intersection(obstacle))!=0:
                    # plt.plot(theta1*m1,theta2*m2,'ko')
                    points.append((theta1,theta2))
                    pass
                pass
            pass
    return points

a_pool = multiprocessing.Pool()
```

```python
t1=time.time()
p=a_pool.map(sum_up_to,range(int(360/m1))) #518400
print(time.time()-t1)
# print(p)
```

```python
#!/usr/bin/env python3
import itertools
import multiprocessing
import math

#Generate values for each parameter
theta1 = range(100)
theta2 = range(100)

#Generate a list of tuples where each tuple is a combination of parameters.
#The list will contain all possible combinations of parameters.
paramlist = list(itertools.product(theta1,theta2))

#A function which will process a tuple of parameters
def func(params):
    theta1 = params[0]
    theta2 = params[1]
    p1=((a0*(math.cos(theta1))),
        (a0*(math.sin(theta1))))
    p2=((a0*(math.cos(theta1))+a1*(math.cos(theta1+theta2))),
        (a0*(math.sin(theta1))+a1*(math.sin(theta1+theta2))))
    links=(Polygon(p0,p1),Polygon(p1,p2))
    # print(theta1,theta2)
    for obstacle in obstacles:
        for link in links:
            if len(link.intersection(obstacle))!=0:
                # plt.plot(theta1*m1,theta2*m2,'ko')
                points.append((theta1,theta2))
                pass
            pass
        pass
    return points

t1=time.time()

#Generate processes equal to the number of cores
pool = multiprocessing.Pool()

#Distribute the parameter sets evenly across the cores
res  = pool.map(func,paramlist)

print(time.time()-t1)
```
